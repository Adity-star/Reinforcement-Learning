# LunarLander-v2 Environment Example

Here we  demonstrates how to use the `LunarLander-v2` environment from the `gymnasium` library, which is a part of the Gym environment toolkit. The `LunarLander-v2` is a physics-based game where an agent must land a spacecraft on the moon using various controls.
- observation: The new state after the action is taken.
- reward: The reward received after taking the action (not used in this code, but could be used to guide learning).
- terminated: A boolean flag that is True if the episode is over (the agent has either succeeded or failed).
- truncated: A boolean flag that indicates if the episode was truncated (cut short, possibly due to a time limit).
- info: Additional data, often left empty but can contain useful information.


```python
# create an evnironment  LunarLander-v2

import gymnasium as gym

env = gym.make("LunarLander-v2")

#Reset the environment
observation, info = env.reset()


for _ in range(20): # Agents interacting 20 times with environment.
  # random action
  action = env.action_space.sample() # returns a random action from a possible actions agent can take
  print("Action taken: ",action)

  observation, reward, terminated, truncated, info = env.step(action)

  if terminated or truncated: # if eposide ended (0/1)
    # reset the environment
    print("Environment is reset")
    observation, info = env.reset()

env.close()
```

# 2. Create the Model
## SB3 PPO Algorithm

*  **Proximal Policy Optimization (PPO)** A policy gradient algorithm that works by maximizing expected reward while ensuring that updates do not move too far away from the previous policy, ensuring training stability
* **MlpPolicy**: This refers to using an MLP (a fully connected neural network) as the agent's policy model. MLPs are simple but effective neural network architectures for many RL problems.

-  we’re going to use SB3 PPO
- PPO is a combination of:

-- Value-based reinforcement learning method: learning an action-value function that will tell us the most valuable action to take given a state and action.

-- Policy-based reinforcement learning method: learning a policy that will give us a probability distribution over actions.

```python
# create environment
env = gym.make('LunarLander-v2')

# initiate the agent
model = PPO('MlpPolicy', env, verbose=1) # agent will use multi layer perceptron to modle policy

# train the agent with 200,000 timesteps.
model.learn(total_timesteps=int(2e5))
```

```python
# SOLUTION
# We added some parameters to accelerate the training
model = PPO(
    policy="MlpPolicy",
    env=env,
    n_steps=1024, # no. of timesteps to collect before updating policy
    batch_size=64, # update policy in 64
    n_epochs=4,  # agent will pass through batch 4 times
    gamma=0.999, # discount factor
    gae_lambda=0.98, # controls tradeoff
    ent_coef=0.01, # penalty
    verbose=1,
)
```

```python
## 3. Training the agent
model_name = "ppo-LunarLander-v2"

# Train it for 1,000,000 timesteps
model.learn(total_timesteps=1000000) # update policy after every batch

# Save the model
model_name = "ppo-LunarLander-v2"
model.save(model_name)
```

- **mean_reward**: The average reward obtained over the 10 evaluation episodes. A higher mean reward indicates that the agent is performing well.

- **std_reward**: The standard deviation of the rewards across the episodes. A smaller standard deviation indicates more consistent performance.

```python
## 4. Evaluate the Agent
eval_env = Monitor(gym.make("LunarLander-v2")) # create a new instance and wrap the agent to track logs

mean_reward, std_reward = evaluate_policy(model, eval_env, n_eval_episodes=10, deterministic=True)
print(f"mean_reward={mean_reward:.2f} +/- {std_reward}")
```
